```python
##### Assembles a shapefile with all sub-national geographies for labor data

import os
import pandas as pd
import geopandas as gpd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_csv(f"{cd}/Data/Clean/Labor/subnational_labor_final.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/subnational_labor_final_geo.shp"
```

```python
### Prep geographies

labor['GEO_ID'] = labor['GEO_ID'].astype('str')

# EU
EU_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/EU_FADN/EU_FADN_2021_regions.shp")
EU_geo['FADN_code_'] = EU_geo['FADN_code_'].astype('str')
EU_geo = EU_geo.merge(labor, left_on='FADN_code_', right_on='GEO_ID', how='inner')

# US
US_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/US_counties/cb_2018_us_county_20m.shp")
US_geo['GEOID'] = US_geo['GEOID'].astype('str')
labor_US = labor.loc[labor['ISO3'] == 'USA'].copy()
labor_US["GEO_ID"] = labor_US["GEO_ID"].astype(str).str.zfill(5)
US_geo = US_geo.merge(labor_US, left_on='GEOID', right_on='GEO_ID', how='inner')

# Canada
CAN_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Canadian_CCS/lccs000a21a_e.shp")
CAN_geo['CCSUID'] = CAN_geo['CCSUID'].astype('str')
CAN_geo = CAN_geo.merge(labor, left_on='CCSUID', right_on='GEO_ID', how='inner')

# China
CHN_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/China_provinces/chn_admbnda_adm1_ocha_2020.shp")
CHN_geo['ADM1_PCODE'] = CHN_geo['ADM1_PCODE'].astype('str')
CHN_geo = CHN_geo.merge(labor, left_on='ADM1_PCODE', right_on='GEO_ID', how='inner')

# Mexico
MEX_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/MEX_state/mex_admin1.shp")
MEX_geo['adm1_pcode'] = MEX_geo['adm1_pcode'].astype('str')
MEX_geo = MEX_geo.merge(labor, left_on='adm1_pcode', right_on='GEO_ID', how='inner')

# Brazil
BRA_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Brazil_Municipalities/BR_Municipios_2023.shp")
BRA_geo['CD_MUN'] = BRA_geo['CD_MUN'].astype('str')
BRA_geo = BRA_geo.merge(labor, left_on='CD_MUN', right_on='GEO_ID', how='inner')

# Ghana
GHA_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Ghana_regions/GHA_regions.shp")
GHA_geo['ISO_1'] = GHA_geo['ISO_1'].astype('str')
GHA_geo = GHA_geo.merge(labor, left_on='ISO_1', right_on='GEO_ID', how='inner')

# India
IND_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/India/India_Districts.shp")
IND_geo['GID_2'] = IND_geo['GID_2'].astype('str')
IND_geo = IND_geo.merge(labor, left_on='GID_2', right_on='GEO_ID', how='inner')

# Turkey
TUR_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Turkey/TUR_regions.shp")
TUR_geo['layer'] = TUR_geo['layer'].astype('str')
TUR_geo = TUR_geo.merge(labor, left_on='layer', right_on='GEO_ID', how='inner')

# South Africa
ZAF_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/South_Africa_municipalities/ZAF_municipalities.shp")
ZAF_geo['CC_3'] = ZAF_geo['CC_3'].astype('str')
ZAF_geo = ZAF_geo.merge(labor, left_on='CC_3', right_on='GEO_ID', how='inner')

# Bolivia
BOL_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Bolivia_municipalities/Bolivia_municipalities.shp")
BOL_geo['GID_3'] = BOL_geo['GID_3'].astype('str')
BOL_geo = BOL_geo.merge(labor, left_on='GID_3', right_on='GEO_ID', how='inner')

# Japan
JPN_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Japan_prefectures/JPN_prefectures.shp")
JPN_geo['HASC_1'] = JPN_geo['HASC_1'].astype('str')
JPN_geo = JPN_geo.merge(labor, left_on='HASC_1', right_on='GEO_ID', how='inner')

# Nigeria
NGA_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Nigeria_states/NGA_states.shp")
NGA_geo['HASC_1'] = NGA_geo['HASC_1'].astype('str')
NGA_geo = NGA_geo.merge(labor, left_on='HASC_1', right_on='GEO_ID', how='inner')

# New Zealand
NZL_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/NewZealand_regions/NZL_regions.shp")
NZL_geo['HASC_1'] = NZL_geo['HASC_1'].astype('str')
NZL_geo = NZL_geo.merge(labor, left_on='HASC_1', right_on='GEO_ID', how='inner')

# Phillipines
PHL_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Philippines_adm_regions/PHL_admin_regions.shp")
PHL_geo['REGION'] = PHL_geo['REGION'].astype('str')
PHL_geo = PHL_geo.merge(labor, left_on='REGION', right_on='GEO_ID', how='inner')

# Russia
RUS_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Russia_regions/Russia_regions.shp")
RUS_geo['HASC_1'] = RUS_geo['HASC_1'].astype('str')
RUS_geo = RUS_geo.merge(labor, left_on='HASC_1', right_on='GEO_ID', how='inner')

# Uganda
UGA_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Uganda/Uganada_subregions.shp")
UGA_geo['ISO_1'] = UGA_geo['ISO_1'].astype('str')
UGA_geo = UGA_geo.merge(labor, left_on='ISO_1', right_on='GEO_ID', how='inner')

# Vietnam
VNM_geo = gpd.read_file("/Users/carinamanitius/Documents/Data/Admin_Boundaries/Vietnam_provinces/Vietnam_provinces.shp")
VNM_geo['HASC_1'] = VNM_geo['HASC_1'].astype('str')
VNM_geo = VNM_geo.merge(labor, left_on='HASC_1', right_on='GEO_ID', how='inner')
```

```python
### Filter geographies to needed data and align CRS

col_to_keep = ['ISO3', 'GEO_ID', 'GEO_ID_NAME', 'ag_jobs', 'geometry']

geo_gdfs = [CAN_geo, CHN_geo, EU_geo, MEX_geo, US_geo, BRA_geo, GHA_geo, TUR_geo, BOL_geo, IND_geo, JPN_geo, NGA_geo, NZL_geo, PHL_geo, RUS_geo, UGA_geo, ZAF_geo, VNM_geo]

for df in geo_gdfs:
    df.drop(columns=[c for c in df.columns if c not in col_to_keep], inplace=True)

# Align CRS
for i, df in enumerate(geo_gdfs):
    if df.crs is None:
        geo_gdfs[i] = df.set_crs(epsg=4326)

geo_gdfs = [df.to_crs(4326) for df in geo_gdfs]
```

```python
### Combine all geodf's and save

full_geo = gpd.GeoDataFrame(
    pd.concat(geo_gdfs, ignore_index=True),
    crs=geo_gdfs[0].crs
)

# final clean to save as SHP
full_geo['ag_jobs'] = full_geo['ag_jobs'].round(0)
full_geo['value'] = full_geo['ag_jobs']
full_geo = full_geo.drop(columns=['ag_jobs'])
full_geo['unit'] = 'ag_jobs'
full_geo = full_geo.rename(columns={'GEO_ID_NAME': 'GEO_ID_NM'})

full_geo.to_file(save_path)
```
