```python
##### Cleans Vietnam labor data
# calculates ag labor and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_excel(f"{cd}/Data/Raw/Sub_National/Vietnam/ag_employment.xlsx")

VNM_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/VNM_provinces.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/VNM_labor.csv"
```

```python
##### Clean

# convert units
labor['2020'] = labor['employed_in_ag_foresty_fishing_thousand_people_2020'] * 1000

# merge with GEO ID
labor = labor.merge(VNM_codes, left_on='Province', right_on='VARNAME_1')

# add units
labor['Units'] = 'Ag labor - jobs'

# keep only needed columns
col_to_keep = ['HASC_1', 'Units', '2020']
labor = labor[col_to_keep]
```

```python
# Save
labor.to_csv(save_path, index=False)
```
