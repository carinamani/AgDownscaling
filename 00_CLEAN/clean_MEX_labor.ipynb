```python
##### Cleans Mexico labor data
# cleans and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_csv(f"{cd}/Data/Raw/Mexico/labor_state.csv")

MEX_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/MEX_state.csv", encoding="latin1")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/MEX_census_labor.csv"
```

```python
##### Clean labor

# keep only agricultre
labor = labor[labor['Sector'] == 'Agriculture, Animal Breeding and Production, Forestry, Fishing and Hunting']

# merge with codes
labor= labor.merge(MEX_codes, on='State_ID', how='outer')

# convert to wide 
labor_wide = labor.pivot(
    index='State_ID',
    columns='Year',
    values='Workforce'  
).reset_index()

# add units
labor_wide['Units'] = 'Ag labor - jobs'
```

```python
# Save cleaned data
labor_wide.to_csv(save_path, index=False)
```
