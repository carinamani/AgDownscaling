```python
##### Cleans Nigeria labor data
# adds formal and informal workers and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_csv(f"{cd}/Data/Raw/Sub_National/Nigeria/NGA_employment.csv")

NGA_states = pd.read_csv(f"{cd}/Data/Correspondence_tables/NGA_states.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/NGA_labor.csv"
```

```python
labor
```

```python
##### Adds formal and informal workers

labor['informal_ag_work'] = labor['informal_employment'] - labor['informal_employment_excluding_ag']

labor['2023'] = labor['informal_ag_work'] + labor['employed_in_ag']
```

```python
##### Clean

# add units
labor['Units'] = 'Ag labor - jobs'

# keep only neccesary columns
col_to_keep = ['HASC_1', 'Units', '2023']
labor = labor[col_to_keep]
```

```python
##### Save
labor.to_csv(save_path, index=False)
```
