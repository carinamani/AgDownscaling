```python
##### Cleans Uganda labor data
# gets average % of employment in each region, aligns with natinoal statistic for 2021, and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_excel(f"{cd}/Data/Raw/Sub_National/Uganda/ag_employment.xlsx")

UGA_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/UGA_subregions.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/UGA_labor.csv"
```

```python
#### Clean

# get average share of labor in each subregion
labor ['average_share'] = (labor ['Share_ag_employment_market'] + labor ['Share_ag_employment_non_market']) / 2

share_sum = labor['average_share'].sum()
labor ['share_of_sum'] = labor ['average_share'] / share_sum

# convert shares to workers using 2021 number from ILO country-wide statistic 
UGA_ag_employment_2021 = 12931560
labor ['2021'] = labor ['share_of_sum'] * UGA_ag_employment_2021

# add units
labor['Units'] = 'Ag labor - jobs'

# merge with GEO ID
labor = labor.merge(UGA_codes, left_on='Sub-region', right_on='name')

# keep only needed columns
col_to_keep = ['ISO_1', 'Units', '2021']
labor = labor[col_to_keep]
```

```python
# Save
labor.to_csv(save_path, index=False)
```
