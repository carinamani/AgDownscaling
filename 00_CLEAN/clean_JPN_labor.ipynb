```python
##### Cleans Japan labor data
# combines across permenant employees, full-time, and temporary workers and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_excel(f"{cd}/Data/Raw/Sub_National/Japan/Japan_ag_employment.xls", skiprows=19, skipfooter=4, engine='xlrd', sheet_name='n146_147')

JPN_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/JPN_prefectures.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/JPN_labor.csv"
```

```python
##### Clean 

# 7 = employees, 12 = permanently hired workers on the farm, 19 = temporary hired workers on the farm
labor['2020'] = labor['Unnamed: 7'] + labor['Unnamed: 12'] + labor['Unnamed: 19']

# merge 
labor = labor.merge(JPN_codes, left_on='Unnamed: 27', right_on='Census_name', how='outer')  

# add units
labor['Units'] = 'Ag labor - jobs'

# keep only relevant columns
col_to_keep = ['HASC_1', 'Units', '2020']
labor = labor[col_to_keep]
```

```python
##### Save
labor.to_csv(save_path, index=False)
```
