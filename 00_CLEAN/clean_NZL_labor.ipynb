```python
##### Cleans New Zealand labor data
# cleans and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_csv(f"{cd}/Data/Raw/Sub_National/New_Zealand/ag_employment.csv")

codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/NZL_regions.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/NZL_labor.csv"
```

```python
##### Clean

# pivot to wide
labor_wide = labor.pivot(
    index='Area', 
    columns='Year', 
    values='OBS_VALUE'
).reset_index()

# add units
labor_wide['Units'] = 'Ag labor - jobs'

# merge with GEO ID
labor_wide = labor_wide.merge(codes, left_on='Area', right_on='Region_Name', how='inner')

# keep only needed columns
col_to_keep = ['HASC_1', 'Units', 2006, 2013, 2018]
labor_wide = labor_wide[col_to_keep]
```

```python
# Save
labor_wide.to_csv(save_path, index=False)
```
