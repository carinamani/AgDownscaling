```python
##### Cleans Turkey capital stock data
# combines machinery and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
capital_stock = pd.read_csv(f"{cd}/Data/Raw/Sub_National/Turkey/capital_stock.csv")

codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/TUR_provinces.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Capital_stock/TUR_capital_stock.csv"
```

```python
##### Clean

# Sum all machines
cols = ['Plough', 'Sowing machine', 'Manure spreading machinery',
        'Water pump', 'Stationary milking plant machine',
        'Mobile milking plant machine', 'Combine harvester', 'Tractor']

capital_stock[cols] = capital_stock[cols].apply(pd.to_numeric, errors='coerce').fillna(0)

capital_stock['2020'] = capital_stock['Plough'].astype(int) + capital_stock['Sowing machine'].astype(int) + capital_stock['Manure spreading machinery'].astype(int) + capital_stock['Water pump'].astype(int) + capital_stock['Stationary milking plant machine'].astype(int) + capital_stock['Mobile milking plant machine'].astype(int) + capital_stock['Combine harvester'].astype(int) + capital_stock['Tractor'].astype(int)

# units
capital_stock['Units'] = 'Ag capital stock - count of tractors, machinery, and equipment'

# merge with ID's
capital_stock = capital_stock.merge(codes, left_on='Region Code', right_on='Region_code', how='outer')

# filter columns
col_to_keep = ['HASC_1', 'Units', '2020']
capital_stock = capital_stock[col_to_keep]
```

```python
# Save
capital_stock.to_csv(save_path, index=False)
```
