```python
##### Cleans Bolivia labor data
# cleans and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_excel(f"{cd}/Data/Raw/Sub_National/Bolivia/Bolivia_employment.xlsx", skiprows=7, skipfooter=2)

BOL_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/BOL_municipalities.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/BOL_labor.csv"
```

```python
##### Clean

# filter down to only data for municipalities 
labor = labor[labor["Unnamed: 5"] == labor["Unnamed: 6"]]

# rename ag columns 
labor = labor.rename(columns={'Unnamed: 10': '2012', 'Unnamed: 79': '2024'})

# add units
labor['Units'] = 'Ag labor - jobs'

# add ID column for merge
labor["id"] = range(1, len(labor) + 1)

# merge with geography codes
labor = labor.merge(BOL_codes, left_on='id', right_on='Census_ID', how='inner')

# keep only relevant columns
col_to_keep = ['GID_3', 'Units', '2012', '2024']
labor = labor[col_to_keep]
```

```python
### Final cleaning 

labor['ISO3'] = 'BOL'
labor['Year'] = 2024
labor['GEO_ID_NAME'] = 'GID_3'

labor = labor.rename(columns={'2024': 'ag_labor_jobs', 'GID_3': 'GEO_ID'})

col_to_keep = ['ISO3', 'GEO_ID', 'GEO_ID_NAME', 'Year', 'ag_labor_jobs']
labor = labor[col_to_keep]
```

```python
# save
labor.to_csv(save_path, index=False)
```
