```python
##### Cleans Tanzania capital stock data
# reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
capital_stock = pd.read_csv(f"{cd}/Data/Raw/Sub_National/Tanzania/households.csv")

TZA_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/TZA_regions.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Capital_stock/TZA_capital_stock.csv"
```

```python
##### CLEAN

capital_stock['ISO3'] = 'TZA'
capital_stock['Year'] = 2020
capital_stock['GEO_ID_NAME'] = 'GID_1'

# merge with region data
capital_stock = capital_stock.merge(TZA_codes, on='GID_1', how='outer')

# calculate # of tractor-owning households
capital_stock['ag_capital_stock_count_machinery'] = capital_stock['percent_households_with_tractor'] * capital_stock['total_agricultural_households']

# rename GEO column
capital_stock.rename(columns={'GID_1': 'GEO_ID'}, inplace=True)

# select columns
col_to_keep = ['ISO3', 'GEO_ID', 'GEO_ID_NAME', 'Year', 'ag_capital_stock_count_machinery']
capital_stock = capital_stock[col_to_keep]
```

```python
##### Save cleaned data
capital_stock.to_csv(save_path, index=False)
```
