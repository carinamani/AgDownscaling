```python
##### Cleans Qatar capital and labor data
# reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
data = pd.read_csv(f"{cd}/Data/Raw/Sub_National/Qatar/ag_census.csv")

QAT_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/QAT_municipalities.csv")

# Set save path
save_path_capital = f"{cd}/Data/Clean/Capital_stock/QAT_capital_stock.csv"
save_path_labor = f"{cd}/Data/Clean/Labor/QAT_labor.csv"
```

```python
### Clean 

# merge with region ID
data = data.merge(QAT_codes, on='GID_1', how='outer')

# rename columns
data.rename(columns={'GID_1': 'GEO_ID'}, inplace=True)

# add variables
data['ISO3'] = 'QAT'
data['Year'] = 2021
data['GEO_ID_NAME'] = 'GID_1'
```

```python
### Clean labor data

labor = data.copy()

labor = labor.rename(columns={'number_of_workers': 'ag_labor_jobs'})

col_to_keep = ['ISO3', 'GEO_ID', 'GEO_ID_NAME', 'Year', 'ag_labor_jobs']
labor = labor[col_to_keep]

##### Save 
labor.to_csv(save_path_labor, index=False)
```

```python
##### Clean capital stock

capital_stock = data.copy()

# add units
capital_stock['Units'] = 'Ag capital stock - count of tractors, machinery, and equipment'

# rename columns
capital_stock.rename(columns={'number_of_tractors': 'ag_capital_stock_count_machinery'}, inplace=True)

col_to_keep = ['ISO3', 'GEO_ID', 'GEO_ID_NAME', 'Year', 'ag_capital_stock_count_machinery']
capital_stock = capital_stock[col_to_keep]

##### Save
capital_stock.to_csv(save_path_capital, index=False)
```
