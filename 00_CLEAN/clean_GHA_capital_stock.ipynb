```python
##### Cleans Ghana capital stock data
# reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
capital_stock = pd.read_csv(f"{cd}/Data/Raw/Sub_National/Ghana/agricultural_equipment.csv")

GHA_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/GHA_regions.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Capital_stock/GHA_capital_stock.csv"
```

```python
##### CLEAN

# add units
capital_stock['Units'] = 'Ag capital stock - count of tractors, machinery, and equipment'

# rename columns
capital_stock.rename(columns={'Use': '2017'}, inplace=True)

# merge with region ID
capital_stock = capital_stock.merge(GHA_codes, right_on='NAME_1', left_on='Region', how='outer')

# keep needed columns
col_to_keep = ['ISO_1', 'Units', '2017']
capital_stock = capital_stock[col_to_keep]
```

```python
### Final cleaning 

capital_stock['ISO3'] = 'GHA'
capital_stock['Year'] = 2017
capital_stock['GEO_ID_NAME'] = 'ISO_1'

capital_stock = capital_stock.rename(columns={'2017': 'ag_capital_stock_count_machinery', 'ISO_1': 'GEO_ID'})

col_to_keep = ['ISO3', 'GEO_ID', 'GEO_ID_NAME', 'Year', 'ag_capital_stock_count_machinery']
capital_stock = capital_stock[col_to_keep]
```

```python
##### Save cleaned data
capital_stock.to_csv(save_path, index=False)
```
