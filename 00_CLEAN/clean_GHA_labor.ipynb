```python
##### Cleans Ghana labor data
# reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_csv(f"{cd}/Data/Raw/Sub_National/Ghana/agricultural_workers.csv")

GHA_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/GHA_regions.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/GHA_labor_census.csv"
```

```python
##### CLEAN

# add units
labor['Units'] = 'Ag labor - jobs'

# rename columns
labor.rename(columns={'Both Sexes': '2017'}, inplace=True)

# merge with region ID
labor = labor.merge(GHA_codes, right_on='NAME_1', left_on='Region', how='outer')

# keep needed columns
col_to_keep = ['ISO_1', 'Units', '2017']
labor = labor[col_to_keep]
```

```python
##### Save cleaned data
labor.to_csv(save_path, index=False)
```
