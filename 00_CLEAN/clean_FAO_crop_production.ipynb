```python
##### Cleans FAO crop production volume (primary products) data
# aggregates into total for each country

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
production = pd.read_csv(f"{cd}/Data/Raw/FAO_production/crop_primary_production.csv")

commodities = pd.read_csv(f"{cd}/Data/Correspondence_tables/FAO_commodity_groups.csv")

country_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/country_names.csv", encoding="cp1252")

# Set save path
save_path = f"{cd}/Data/Clean/Production/FAO_data/FAO_crop_production.csv"
```

```python
#### Merge with country codes

production = production.merge(country_codes, left_on='Area Code (M49)', right_on='FAO_code', how='left')

# drop if missing ISO code
production = production.dropna(subset=['ISO3'])
```

```python
##### Get sum of tonne values

production_total = production.groupby('ISO3')['Value'].sum().reset_index()
```

```python
##### Save

production_total.to_csv(save_path, index=False)
```
