```python
##### Cleans Turkey labor data
# cleans and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_csv(f"{cd}/Data/Raw/Sub_National/Turkey/ag_employment.csv", encoding="cp1254")

codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/TUR_regions.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/TUR_labor.csv"
```

```python
##### Clean

# Convert jobs to 0's
labor['2020'] = labor['Agriculture (1000)'] * 1000

# add units
labor['Units'] = 'Ag labor - jobs'

# merge with GEO ID
labor = labor.merge(codes, left_on='Region Code', right_on='layer', how='outer')

# keep only neccesary columns
col_to_keep = ['layer', 'Units', '2020']
labor = labor[col_to_keep]
```

```python
# Save
labor.to_csv(save_path, index=False)
```
