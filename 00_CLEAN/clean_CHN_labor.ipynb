```python
##### Cleans China labor data
# changes units and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_csv(f"{cd}/Data/Raw/CHN_stats/Agriculture_forestry_fishing_employment_10000persons.csv")

CHN_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/CHN_provinces.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/CHN_province_labor_NBS.csv"
```

```python
##### Clean

# merge to get admin codes
labor = labor.merge(CHN_codes, left_on='Region', right_on='STAT_name', how='outer')

# convert to total jobs
year_cols = [str(year) for year in range(2005, 2025)]
labor[year_cols] = labor[year_cols] * 10_000

# add units
labor['Units'] = 'Ag labor - jobs'

# reorder columns
columns_to_keep = ['ADM_CODE', 'Units', '2024', '2023', '2022', '2021', '2020', '2019', '2018',
       '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2010', '2009',
       '2008']
labor = labor[columns_to_keep]
```

```python
##### Save cleaned data
labor.to_csv(save_path, index=False)
```
