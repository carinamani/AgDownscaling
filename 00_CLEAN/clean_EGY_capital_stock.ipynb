```python
##### Cleans Egypt capital stock data
# cleans and reformats 

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
capital_stock = pd.read_csv(f"{cd}/Data/Raw/Sub_National/Egypt/EGY_ag_machines.csv", skiprows=1)

EGY_gov = pd.read_csv(f"{cd}/Data/Correspondence_tables/EGY_governorates.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Capital_stock/EGY_capital_stock.csv"
```

```python
##### Clean data

# merge with ID's
capital_stock = capital_stock.merge(EGY_gov, left_on='Indicator', right_on='STATS_NAME', how='outer')

# units
capital_stock['Units'] = 'Ag capital stock - count of tractors, machinery, and equipment'

# filter columns
col_to_keep = ['HASC_1', 'Units', '2011', '2013', '2015', '2017', '2019', '2021', '2023']

capital_stock = capital_stock[col_to_keep]

# drop NA
capital_stock = capital_stock.dropna()
```

```python
### Final cleaning 

capital_stock['ISO3'] = 'EGY'
capital_stock['Year'] = 2021
capital_stock['GEO_ID_NAME'] = 'HASC_1'

capital_stock = capital_stock.rename(columns={'2021': 'ag_capital_stock_count_machinery', 'HASC_1': 'GEO_ID'})

col_to_keep = ['ISO3', 'GEO_ID', 'GEO_ID_NAME', 'Year', 'ag_capital_stock_count_machinery']
capital_stock = capital_stock[col_to_keep]
```

```python
##### Save
capital_stock.to_csv(save_path)
```
