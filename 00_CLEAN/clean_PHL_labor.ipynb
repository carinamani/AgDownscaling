```python
##### Cleans Philippines labor data
# calculates ag labor and reformats

import os
import pandas as pd
```

```python
##### Load data

# Get the current working directory
cd = os.path.dirname(os.getcwd())

# Import data
labor = pd.read_excel(f"{cd}/Data/Raw/Sub_National/Philippines/PHL_ag_employment.xlsx")

PHL_codes = pd.read_csv(f"{cd}/Data/Correspondence_tables/PHL_admin_regions.csv")

# Set save path
save_path = f"{cd}/Data/Clean/Labor/PHL_labor.csv"
```

```python
##### Clean

# Calculate labor 
labor['2024'] = labor['total_workers_thousands_2024'] * (labor['share_ag_and_forestry_2024']/100) * 1000

# add units
labor['Units'] = 'Ag labor - jobs'

# merge with GEO ID
labor = labor.merge(PHL_codes, left_on='region', right_on='Stats_Name', how='inner')

# keep only needed columns
col_to_keep = ['REGION', 'Units', '2024']
labor = labor[col_to_keep]
```

```python
# Save
labor.to_csv(save_path, index=False)
```
